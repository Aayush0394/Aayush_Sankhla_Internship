```python
import numpy as np
#Replacing NaN with 0 
arr=np.array([[6, -8, 73, -110], [np.nan, -8, 0, 94]])
np.nan_to_num(arr,nan=0,copy=False)
print(arr)
#Interchanging rows and columns
arr[[0,1],:]=arr[[1,0],:]
print("After interchanging rows:\n",arr)
arr[:,[0,1,2]]=arr[:,[2,0,1]]
print("After interchanign columns:\n",arr)
```

```python
#Moving axes of  a 3D array
threeD=np.full((2,3,2),1)
print("Moving axis")
print(threeD.shape)
res=np.moveaxis(threeD,1,2)
print(res.shape)
```

```python
#Replacing NaN values with column average
org=np.array([[1,2,3,np.nan],[2,np.nan,2,4]])
for col in range(org.shape[1]):
    mean = np.nanmean(org[:, col])

    for row in range(org.shape[0]):
        if np.isnan(org[row, col]):
            org[row, col] = mean
print("Relaced NaN with mean")
print(org)
```

```python
#Relacing -ve values
arr=np.array([23,-3,3,1,-14,8])
arr[(arr < 0)]=0
print(arr)
```

```python
#create a numpy 1d-arrays 
arr1 = np.array([3, 4]) 
arr2 = np.array([1, 0]) 
# find average of NumPy arrays avg = (arr1 + arr2) / 2 print("Average of NumPy arrays:\n", avg) 
avg=(arr1+arr2)/2
print("Average of Numpy Arrays:\n",avg)
```

```python
#Calculating mean , median and mode
comb=np.concatenate((arr1,arr2))
print("Mean:",np.mean(comb))
print("Median:",np.median(comb))
#Finding mode 
max_count = 0
mode = None
for i in range(len(comb)):
    count = 0
    for j in range(len(comb)):
        if comb[i] == comb[j]:
            count += 1
    if count > max_count:
        max_count = count
        mode = comb[i]
print("Mode:", mode)
```

```python
#Solving linear equation
# x - 2y + 3z = 9 -x + 3y - z = -6 2x - 5y + 5z = 17
A=[[1,-2,2],[-1,3,-1],[2,-5,5]]
B=[9,-6,17]
A=np.array(A)
B=np.array(B)
x,y,z=np.linalg.solve(A,B)
print("X:",x,"Y:",y,"Z:",z)
```

```python
#Inverse method
Ainv=np.linalg.inv(A)
res=np.dot(Ainv,B)
print("Result:",res)
```
