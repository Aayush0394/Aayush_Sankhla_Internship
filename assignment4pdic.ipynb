```python
#Repeating tuple
tp =(2,3)*3
print(tp,type(tp))
```

```python
#Joining tuples using + operator
tp1=(1,2)
tp2=(3,4)
tp3=(5,6)
finalTp=tp1+tp2+tp3
print(finalTp,type(finalTp))
```

```python
#Finding element using in keyword
words=("hai","hello","hmm","bye","namastey")
if "hai" in words:
    print("'hai' exist in words")
else:
    print("'hai' is not present in words")
```

```python
#Calculating total , highest , lowest values in tuple without in-built methods
num=(1,4,3,5,2)
n=len(num)
highest=num[0]
lowest=num[0]
total=0
for i in num:
    total+=i
    if i<lowest:
        lowest=i
    if i>highest:
        highest=i
print("Total:",total)
print("Lowest:",lowest)
print("Highest:",highest)
```

```python
#Filtering tuple
n = (3, 14, 7, 22, 9, 41, 18, 5)
ans=tuple(x for x in n if x > 10)
print("Filtered Tuple:",ans)
```

```python
#Finding length without using len()
s = {"cat", "dog", "bird", "fish"}
length=0
for x in s:
    length+=1
print("No. of items in set:",length)
```

```python
#Finding all unique elements from two sets
s1={1,4,5,6,2}
s2={3,2,7,8,4}
ans=s1.union(s2)
print("Unique elements:",ans)
```

```python
#Finding all common elements from two sets
s1 = {1, 2, 3, 4} 
s2 = {3, 4, 5, 6}
ans=s1.intersection(s2)
print("Common elements:",ans)
```
